# TITANIC CLASSIFICATION

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# Load the Titanic dataset
df = pd.read_csv('Titanic-Dataset.csv')

```

```python
# Display the first few rows of the dataframe
print(df.head())

```

```python
# Display basic statistics of the dataframe
print(df.describe())
```

```python
# Display the info of the dataframe to understand the data types and missing values
print(df.info())
```

### The Visualizations showing the survival rates based on different factors:

```python
# Set the background color to white for better visibility
plt.figure(facecolor='white')

# Plot the survival rate by socio-economic status (Pclass) with custom colors
sns.countplot(x='Pclass', hue='Survived', data=df, palette='Set2')
plt.title('Survival Rate by Socio-Economic Status (Pclass)')
plt.xlabel('Pclass')
plt.ylabel('Count')
plt.show()
```

```python
# Plot the survival rate by gender with custom colors
plt.figure(facecolor='white')
sns.countplot(x='Sex', hue='Survived', data=df, palette='Set3')
plt.title('Survival Rate by Gender')
plt.xlabel('Gender')
plt.ylabel('Count')
plt.show()
```

```python
# Plot the survival rate by age with custom colors
plt.figure(facecolor='white')
sns.histplot(data=df, x='Age', hue='Survived', multiple='stack', kde=True, palette='Set4')
plt.title('Survival Rate by Age')
plt.xlabel('Age')
plt.ylabel('Count')
plt.show()
```

```python
# Plot the survival rate by fare
plt.figure(facecolor='white')
sns.histplot(data=df, x='Fare', hue='Survived', multiple='stack', kde=True)
plt.title('Survival Rate by Fare')
plt.xlabel('Fare')
plt.ylabel('Count')
plt.show()
```
